```python
# sets don't allow duplicate values
spanish_stopwords = set(stopwords.words('spanish'))
custom_stop_words = ['contributors', 'contributor', 'translations', 'lyrics', 'you', 'might', 'also', 'like', 'embed', 'english']
spanish_stopwords.update(custom_stop_words)
```

```python
df.lyrics[32]
```

```python
def clean_lyrics(lyrics):

   # clean [.*] and (.*) characters
   stage_1 = sub('\[.+?\]|\(.+?\)','',lyrics)

   # clean numbers
   stage_2 = sub('\d+','',stage_1)

   # clean punctuation
   stage_3 = sub('\,+|\.+|\-+|\*|\¿+|\?+|\!+|\¡+|\"+|\:+','',stage_2)

   # space where uppercase and lowercase are together
   # \1 and \2 are matching groups for 2 () () respectively
   stage_4 = sub('([a-z])([A-Z])',r'\1 \2',stage_3)
   
   # clean jumplines
   stage_5 = sub('\n+',' ',stage_4)

   # all words in lowercase
   stage_6 = stage_5.lower().strip()
   
   # split 
   stage_7 = stage_6.split()
   
   # remove stopwords
   stage_8 = stage_7.copy()
   for word in stage_7:
      stage_8.remove(word) if word in spanish_stopwords else None

   stage_8 = ' '.join(stage_8)
   
   return stage_8
```

```python
df['clean_lyrics'] = df.lyrics.apply(clean_lyrics)
#clean_lyrics(df.lyrics[32])
```
